Formulas for first derivative:

Forward: $f'(x)\approx \frac{f(x+\Delta x)-f(x)}{\Delta x}$

Backward: $f'(x)\approx \frac{f(x)-f(x-\Delta x)}{\Delta x}$

Centered: $f'(x)\approx \frac{f(x_{i+1})-f(x_{i-1})}{2\Delta x}$

- For $\Delta x$ use the values $\{ 2^{-n}, n=1,\ldots 13\}$



```python
# The following code finds the forward, backward, and centered difference
# derivative approximations for a given function for a range of delta 
# values, then plots the absolute errors on a log-log scale plot.

# "Magic" statement for jupyter notebook to display plots ( 1 line)
%matplotlib inline

# Necessary imports (2 lines)
import numpy as np
import matplotlib.pyplot as plt


### Initialization section

# Define function  that is being differentiated (give it the name `fun`)
def fun(x):
    return np.exp(x)

# Point where derivative is evaluated (use the variable x)
x = 0.5

#get vector of Delta x values (call it deltax )
deltax = np.array([2**-1, 2**-2, 2**-3, 2**-4, 2**-5, 2**-6, 2**-7, 2**-8, 2**-9, 2**-10, 2**-11, 2**-12, 2**-13])


# Define the actual derivative of the function being calculated (you have
# to do this by hand).  Give it the name `dfun`
dfun = np.exp(x) * np.ones(len(deltax))



### Calculation section
# Compute forward approximations. Your result should be a vector of the same
# length as deltax.  Do NOT use a loop--instead, use vector calculations.
forward = (fun(x+deltax) - fun(x))/deltax


# Compute backward approximations. The same comments apply as in forward approx.
backward = (fun(x) - fun(x-deltax))/deltax

# Compute centered approximations as the average of forward and backward approx.
centered = (0.5*forward) + (0.5*backward)

# Compute absolute errors for forward, backward, and centered approximations.
# (This will give three separate vectors)
forward_abs_error = abs(dfun - forward) 
backward_abs_error = abs(dfun - backward)
centered_abs_error = abs(dfun - centered)



### Display section
# Create figure 1
fig = plt.figure(figsize = (10,7))


# plot forward, backward, and centered errors on log-log scale, all on the 
# same plot.  Make sure that the x values are increasing as you move right 
# on the x axis.  You may have to google to find out how to do this.
ax = fig.add_subplot(111)
ax.loglog(deltax, forward_abs_error, 'r-', label = 'forward', linewidth = 3)
ax.loglog(deltax, backward_abs_error, 'g-.', label = 'backward', linewidth = 3)
ax.loglog(deltax, centered_abs_error, 'bD-.', label = 'centered', linewidth = 3)


# Label axes. Add legend and title.

ax.set_xlabel('$\Delta x$', fontsize = 12)
ax.set_ylabel('error', fontsize = 12)
ax.set_title('Error vs Changing Delta ($f(x) = e^{x}$)')

ax.legend()



###NOTES
### WAS TRYING TO DO IT IN ONE LINE OF CODE BUT HAD ISSUES WITH THE LEGEND (see code below)
# ax.loglog(deltax, forward_abs_error, 'r-', deltax, backward_abs_error, 'g-.', deltax, centered_abs_error, 'b*')
# ax.legend([forward_abs_error, backward_abs_error, centered_abs_error], ['forward', 'backward', 'centered'], loc = 0)

# could we have used a for loop? perhaps!!

```
